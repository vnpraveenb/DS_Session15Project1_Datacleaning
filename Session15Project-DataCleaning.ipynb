___12. Join users to transactions, keeping all rows from transactions and only matching
rows from users (left join)___ 


```python
# Merging transactions and users using left join
left_join_df = pd.merge(transactions,users,on='UserID', how= 'left')

# Setting the columns with dates to datetime
left_join_df['Registered'] = pd.to_datetime(left_join_df['Registered'])
left_join_df['Cancelled'] = pd.to_datetime(left_join_df['Cancelled'])
left_join_df['TransactionDate'] = pd.to_datetime(left_join_df['TransactionDate'])

left_join_df
```

___14. Join users to transactions, keeping only rows from transactions and users that
match via UserID (inner join)___

```python
inner_join_df = pd.merge(transactions,users,on='UserID', how= 'inner')

# Setting the columns with dates to datetime
inner_join_df['Registered'] = pd.to_datetime(inner_join_df['Registered'])
inner_join_df['Cancelled'] = pd.to_datetime(inner_join_df['Cancelled'])
inner_join_df['TransactionDate'] = pd.to_datetime(inner_join_df['TransactionDate'])
inner_join_df
```

___15. Join users to transactions, displaying all matching rows AND all non-matching rows
(full outer join)___

```python
outer_join_df = pd.merge(transactions,users, how= 'outer')

# Setting the columns with dates to datetime
outer_join_df['Registered'] = pd.to_datetime(outer_join_df['Registered'])
outer_join_df['Cancelled'] = pd.to_datetime(outer_join_df['Cancelled'])
outer_join_df['TransactionDate'] = pd.to_datetime(outer_join_df['TransactionDate'])

outer_join_df
```

___16. Determine which sessions occurred on the same day each user registered___

```python
user_session_df = pd.merge(users,sessions,on='UserID', how= 'inner')

# Setting the columns with dates to datetime
user_session_df['Registered'] = pd.to_datetime(user_session_df['Registered'])
user_session_df['Cancelled'] = pd.to_datetime(user_session_df['Cancelled'])
user_session_df['SessionDate'] = pd.to_datetime(user_session_df['SessionDate'])

user_session_df[user_session_df.Registered == user_session_df.SessionDate]
```

___17. Build a dataset with every possible (UserID, ProductID) pair (cross join)___

```python
userID_df = pd.DataFrame({'key': np.repeat(1, users.shape[0]), 'UserID': users.UserID})
productID_df = pd.DataFrame({'key': np.repeat(1, products.shape[0]), 'ProductID': products.ProductID})

userID_productID_df = pd.merge(userID_df, productID_df,on='key')[['UserID', 'ProductID']]

userID_productID_df
```

___18. Determine how much quantity of each product was purchased by each user___

```python
pd.merge(userID_productID_df, transactions, how='left', on=['UserID', 'ProductID']).groupby(['UserID', 'ProductID']).apply(lambda x: pd.Series(dict(Quantity=x.Quantity.sum()))).reset_index().fillna(0)
```

___20. Join each user to his/her first occuring transaction in the transactions table___

```python
data = pd.merge(users, transactions.groupby('UserID').first().reset_index(), how='left', on='UserID')

# Setting the columns with dates to datetime
data['Registered'] = pd.to_datetime(data['Registered'])
data['Cancelled'] = pd.to_datetime(data['Cancelled'])
data['TransactionDate'] = pd.to_datetime(data['TransactionDate'])

data
```

```python
list(data.dropna(thresh=int(data.shape[0] * .9), axis=1).columns) #set threshold to drop NAs
```

```python
missing_info = list(data.columns[data.isnull().any()])
missing_info
```

```python
for col in missing_info:
    num_missing = data[data[col].isnull() == True].shape[0]
    print('number missing for column {}: {}'.format(col, num_missing)) #count of missing data
```

```python
for col in missing_info:
    percent_missing = data[data[col].isnull() == True].shape[0] / data.shape[0]
    print('percent missing for column {}: {}'.format(col, percent_missing))
```
